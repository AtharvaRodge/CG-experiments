```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Read the dataset
df = pd.read_csv("/iris.csv")

# Display the first few rows
print(df.head())

# Display the last few rows
print(df.tail())

# Display summary statistics
print(df.describe())

# Calculate sum, mean, median, and mode of "sepal.length"
sum_data = df["sepal_length"].sum()
mean_data = df["sepal_length"].mean()
median_data = df["sepal_length"].median()
mode_data = df["sepal_length"].mode()

# Print the results
print("Sum:", sum_data)
print("Mean:", mean_data)
print("Median:", median_data)
print("Mode:", mode_data)
```
